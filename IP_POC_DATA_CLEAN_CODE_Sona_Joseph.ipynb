## dropping certain column with missing value

```python
# dropping certain column with missing value
print("\n\ndropping certain column with missing value")
main_cols = []
imp = int(input("Number of important columns with null value : "))
for x in range(imp):
    i = input()
    main_cols.append(i)

df.dropna(subset=main_cols, inplace=True)
```

```python
df.isnull().sum()
```

###  impute missing values with mean of column of int type

```python
# impute missing values with mean of column
print("\n\nimputed missing values with mean of column")
cols = df.select_dtypes([np.number]).columns
int_mean = df.select_dtypes([np.number]).mean()
for i in cols:
    df[i].fillna(int_mean[i], inplace=True)
```

### Change datatype to integer

```python
#Change datatype to integer
print("\n\nChange datatype to integer ")
int_list = []
int_dtype = int(input("How many columns do you want to make float64 : "))
for x in range(int_dtype):
    i = input()
    int_list.append(i)
    

# Removing extra unwanted symbols


for x in int_list:
        df[x] = df[x].str.replace('>', '')
        df[x] = df[x].str.replace('<', '')    
    
def test_apply(x):
    for i in x:
        try:

            return float(x)
        except ValueError:
            return 0
    
df[x] = df[x].apply(test_apply)
```
